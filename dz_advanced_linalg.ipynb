### Задание 1

Дан объект в $2D$ пространстве

````python

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````python
# Уменьшение объекта в 2 раза

math_less = np.array([[0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 0.5]])
B = np.array(A)@math_less
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
# Поворот объекта на 130 градусов

math_turn = np.array([[math.cos(130), -math.sin(130),0],
    [math.sin(130), math.cos(130), 0],
    [0, 0, 1]])
B = np.array(A)@math_turn
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
# Отражение объекта относительно оси x=y

mat_zerk = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
B = np.array(A).dot(mat_zerk)
B = np.array(A)@mat_zerk
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````
