```python
import copy

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def key_by_value(dict1: dict, value):
    sum1=0
    for k, v in dict1.items():
        if value in v:
            sum1=sum1+1
            return k
    if sum1==0:
        print('ВОТ ТУТ')
        return 'Такого значения нет в словаре'


def change_key_value(dict1: dict) ->dict:
    dict2=dict()
    for k, v in dict1.items():
        for s in v:
            dict2[s]=k
    return dict2

def name_by_number(dict1: dict, number1):
    if dict1['number']==number1:
        return dict1['name']  
            
def command_p(documents=documents):
    number1 = input('Введите номер документа')
    for i in documents:
        name1 = name_by_number(i, number1)
        if name1:
            return f'Владелец документа: {name1}'
    return 'Документ не найден в базе'

def command_s(directories=directories):
    num_doc = input('Введите номер документа')
    for polka, docs in directories.items():
        if num_doc in docs:
            return f'Документ хранится на полке: {polka}'
    return f'Документ не найден в базе'

def current_list_of(directories=directories):
    print(f'Текущий перечень полок: ', end='')
    print(*[k for k in directories.keys()], sep=', ', end='\n')

def command_abc(directories=directories):
    n = input('Введите номер полки: ')
    if n in directories.keys():
        print(f'Такая полка уже существует.', end=' ')
        current_list_of()
    else:
        directories[n]=list()
        print(f'Полка добавлена.', end=' ')
        current_list_of()

def command_ds(directories=directories, documents=documents):
    n = input('Введите номер полки: ')
    if n in directories:
        if len(directories[n])==0:
            del directories[n]
            print(f'Полка удалена.', end=' ')
            current_list_of()

        else:
            print(f'На полке есть документы, удалите их перед удалением полок.', end=' ')
            current_list_of()
    else:
        print(f'Такой полки не существует.', end=' ')
        current_list_of()

def get_full_information(documents=documents, directories=directories):
    print('Текущий список документов:')
    for doc in documents:
        print(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {key_by_value(directories, doc['number'])}")

def command_ad(documents=documents, directories=directories):
    num_doc = input('Введите номер документа')
    type_doc = input('Введите тип документа')
    name_doc = input('Введите владельца документа')
    polka = input('Введите полку для хранения')
    if polka in directories.keys():
        directories[polka]=directories[polka]+[num_doc]
        documents.append(
        {
        'type': type_doc,
        'number': num_doc, 
        'name': name_doc
        }
        )
        print('Документ добавлен.', end=' ')
    else:
        print('Такой полки не существует. Добавьте полку командой as.')
    get_full_information(documents=documents, directories=directories)

def command_d(documents=documents, directories=directories):
    num_doc = input('Введите номер документа')
    if num_doc not in list(sum(directories.values(), [])):
        print('Документ не найден в базе')
    else:
        for i in documents:
            if i['number']==num_doc:
                del documents[documents.index(i)]
                list1 = copy.deepcopy(directories[change_key_value(directories)[num_doc]])
                list1.remove(num_doc)
                directories[change_key_value(directories)[num_doc]] = list1
                print('Документ удалён.')
    get_full_information(documents=documents, directories=directories)
    
def command_m(documents=documents, directories=directories):
    num_doc = input('Введите номер документа:')
    polka = input('Введите номер полки:') 
    if polka not in directories.keys():
        print('Такой полки не существует')
    else:
        list1 = copy.deepcopy(directories[change_key_value(directories)[num_doc]])
        list1.remove(num_doc)
        directories[change_key_value(directories)[num_doc]] = list1
        directories[polka]=directories[polka]+[num_doc]
    get_full_information(documents=documents, directories=directories)

def main_func():
    command = 'i'
    while command != 'q':
        command = input('Введите команду ')
        match command:
            case 'p':
                if command_p:
                    print(command_p())
            case 's':
                print(command_s())
            case 'l':
                get_full_information()
            case 'abc':
                command_abc()
            case 'ds':
                command_ds()
            case 'ad':
                command_ad()
            case 'd':
                command_d()
            case 'm':
                command_m()
```

```python
main_func()
```
